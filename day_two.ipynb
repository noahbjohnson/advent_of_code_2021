```python
import requests
import config
```

```python
inputData = requests.get("https://adventofcode.com/2021/day/2/input",
                         headers={"Cookie": config.cookie}).text.splitlines()
```

```python
tuples = [[x, int(y)] for x, y in [x.split() for x in inputData]]
```

```python
horizontal = sum([x[1] for x in tuples if x[0] == "forward"])
depth = sum([x[1] for x in tuples if x[0] == "down"]) - sum([x[1] for x in tuples if x[0] == "up"])
```

```python
horizontal * depth
```

```python
aim = 0
depth = 0
horizontal = 0
for x, y in tuples:
    if x == "down":
        aim += y
    elif x == "up":
        aim -= y
    else:
        horizontal += y
        depth += aim * y
```

```python
depth * horizontal
```
