```python
import requests

import config
```

```python
inputData = [
    int(x) for x in
    requests.get("https://adventofcode.com/2021/day/1/input", headers={"Cookie": config.cookie}).text.splitlines()
]
```

```python
sum([
    inputData[i] < inputData[i + 1] for i in range(len(inputData) - 1)
])
```

```python
sum([
    sum(inputData[i:i + 3]) < sum(inputData[i + 1:i + 4]) for i in range(len(inputData) - 2)
])
```
